```python
import pandas as pd
```

### Creating a DataFrame from a dictionary

```python
person = {
    "first": ["Corey", "Jane", "John"],
    "last": ["Schafer", "Doe", "Doe"],
    "email": ["coreymschafer@gmail.com", "janedoe@email.com", "johndoe@email.com"]
}
```

```python
df = pd.DataFrame(person)
```

```python
df
```

```python
filtr = (df["last"] == "Doe")
```

> ***This is a filter mask - a series of Boolean***

```python
filtr
```

```python
df[filtr]
```

> ***Filter applied to the DataFrame***

```python
df.loc[filtr, "email"]
```

### Logical operations
- #### `&` - AND operator

```python
filtr = (df["last"] == "Doe") & (df['first'] == "John")
```

```python
df.loc[filtr, "email"]
```

- #### `|` - OR operator

```python
filtr = (df["last"] == "Schafer") | (df['first'] == "John")
```

```python
df.loc[filtr, "email"]
```

- #### `~` - NOT operator

```python
df.loc[~filtr, "email"]
```
