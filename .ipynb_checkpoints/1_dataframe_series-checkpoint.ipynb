```python
import pandas as pd
```

#### Creating a DataFrame from a dictionary

```python
person = {
    "first": ["Corey", "Jane", "John"],
    "last": ["Schafer", "Doe", "Doe"],
    "email": ["coreymschafer@gmail.com", "janedoe@email.com", "johndoe@email.com"]
}
```

```python
person['email']
```

```python
df = pd.DataFrame(person)
```

```python
df
```

### Accessing a column of a DataFrame

```python
df['email']
```

> ***When accessing a single column, a *Series* is returned***

```python
type(df['email'])
```

### Accessing multiple columns

```python
df[["last", "email"]]
```

> ***When accessing multiple columns, a *DataFrame* is returned***

```python
df
```

```python
df[["email", "last"]]
```

```python
df.columns
```

### Accessing rows
- #### `iloc[]` - only integer values can be specified

```python
df.iloc[0]
```

> ***A Series is returned. The index used here are the column names***

```python
type(df.iloc[0])
```

### Accessing multiple rows

```python
df.iloc[[0, 1]]
```

> ***A DataFrame is returned***

### Accessing particular rows and columns

```python
df.columns
```

```python
df.iloc[[0, 1], [1, 2]]
```

```python
df.iloc[[0, 1], 2]
```

- #### `loc[]` - integers and index labels can be specified

```python
df
```

```python
df.loc[0]
```

```python
df.loc[[0, 1], 'email']
```

```python
df.loc[[0, 1], ['email', 'last']]
```
