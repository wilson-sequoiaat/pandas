```python
import pandas as pd
```

```python
dtype_spec = {
    "upc": str
}
```

```python
df_prev = pd.read_csv("dump/label_insight_master_dump_preprocessed_2024-05-01_09-24-34.csv", dtype=dtype_spec)
df_latest = pd.read_csv("dump/label_insight_master_dump_preprocessed_2024-05-31_15-17-22.csv", dtype=dtype_spec)
```

```python
req_cols = [
    'upc',
    'brand',
    'productTitle',
    'productSize',
    'manufacturer',
    'sizes',
    'categorization.productType',
    'categorization.aisle',
    'categorization.shelf',
    'categorization.category',
    'ingredients.declaration',
    'imageSection.images',
]
```

```python
df_latest_min = df_latest[req_cols]
df_prev_min = df_prev[req_cols]
```

```python
rows_to_keep = []
for upc in df_latest_min['upc'].to_list()[:1000]:
    try:
        for col in df_latest_min:
            latest_val = df_latest_min.loc[df_latest_min['upc'] == upc, col].values[0]
            existing_val = df_prev_min.loc[df_prev_min['upc'] == upc, col].values[0]
    
            if (pd.isna(latest_val) and pd.isna(existing_val)):
                continue
            if (latest_val != existing_val):
                rows_to_keep.append(upc)
                break
    except Exception as e:
        print(e)
```
